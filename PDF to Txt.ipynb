```python
import PyPDF2                                       # pip install pyPdf2

pdf_File = open("Testing.pdf",'rb')                 # Open the pdf file

pdfReader = PyPDF2.PdfFileReader(pdf_File)          # It will read the pdf file

x = pdfReader.numPages                              # It will calculate the Number of pages in the pdf


pageObj = pdfReader.getPage(x-1)                    # It will selects only those number of pages which are provided

text = pageObj.extractText()                        # It will Extract the text from the given pdf


file1 = open("Text.txt","a")                         
file1.writelines(text)
file1.close()
```
